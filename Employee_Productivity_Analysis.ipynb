```python
import pandas as pd
```

```python
data = pd.read_csv('employee_productivity_50rows.csv')
data.head()
```

```python
data.info()
data.describe()
```

```python
data.isnull().sum()
```

```python
data.info()
```

```python

data['Department'] = data['Department'].astype('category').cat.codes
data['Remote'] = data['Remote'].map({'Yes': 1, 'No': 0})
data['Promotions'] = data['Promotions'].map({'Yes': 1, 'No': 0})


data.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
plt.show()
```

```python
sns.scatterplot(x='Hours_Worked', y='Tasks_Completed', hue='Department', data=data)
plt.title("Hours Worked vs Tasks Completed")
plt.show()
```

```python
sns.barplot(x='Department', y='Tasks_Completed', data=data)
plt.title("Average Tasks Completed per Department")
plt.show()
```

```python
sns.boxplot(x='Remote', y='Engagement_Score', data=data)
plt.title("Engagement Score: Remote vs On-site")
plt.show()
```
